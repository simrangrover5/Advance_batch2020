```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C://internship/datascience/DS_Assignments/missing_migrants/MissingMigrantsGlobal.csv",encoding="latin",sep="\t")
```

```python
data.head()
```

```python
#What is the percentage of missing females, males and children in top 10 region of incidents??
```

```python
total = data.groupby("Region of Incident").count().sort_values(by="Web ID",ascending=False)[:10]["Number of Females"].sum()
```

```python
data.groupby("Region of Incident").count().sort_values(by="Web ID",ascending=False)[:10]["Number of Females"]/total
```

```python
for c in ["Number of Females","Number of Males","Number of Children"]:
    total = data.groupby("Region of Incident").count().sort_values(by="Web ID",ascending=False)[:10][c].sum()
    print(data.groupby("Region of Incident").count().sort_values(by="Web ID",ascending=False)[:10][c]/total)
```

```python
for c in ["Number of Females","Number of Males","Number of Children"]:
    total = data.groupby("Region of Incident").count().sort_values(by="Web ID",ascending=False)[:10][c].sum()
    (data.groupby("Region of Incident").count().sort_values(by="Web ID",ascending=False)[:10][c]/total).plot(kind="bar",title=c)
    plt.show()
```
