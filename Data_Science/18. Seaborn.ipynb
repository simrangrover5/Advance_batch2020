```python
fig = plt.figure(facecolor="#abcdef")
ax = fig.add_axes([0,0,1,1],facecolor="#abcdef")
plt.boxplot([tip['total_bill'],tip['size']],showmeans=True,flierprops={"markeredgecolor":"red","markerfacecolor":"black"})
plt.yticks(range(1,50,3),fontsize=14,color="white")
plt.xticks(color="white")
plt.show()
```

```python
fig = plt.figure(facecolor="#abcdef")
ax = fig.add_axes([0,0,1,1],facecolor="#abcdef")
plt.yticks(range(1,50,3),fontsize=14,color="white")
plt.xticks(color="white")
a = plt.boxplot([tip['total_bill'],tip['size']],showmeans=True,flierprops={"markeredgecolor":"red","markerfacecolor":"black"})
for b in a['boxes']: 
    b.set_linewidth(3)
    b.set_linestyle("--")
for m in a['means']:
    m.set_markerfacecolor("black")
    m.set_markeredgecolor("black")
for w in range(len(a['whiskers'])):
    if w%2 == 0:
        a['whiskers'][w].set_linestyle(":")
        a['whiskers'][w].set_linewidth(4)
    else:
        a['whiskers'][w].set_linestyle("-.")
        a['whiskers'][w].set_linewidth(4)
    a['whiskers'][w].set_color("#bca123")
```
