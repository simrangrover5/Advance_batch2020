```python
class Mysearch(StreamListener):
    def on_data(self,tweet):
        f = open("tweets.txt","a")
        f.write(tweet)
        f.write("\n")
        f.close()
    
obj = Mysearch()
f = Stream(auth,obj)
f.filter(track=['corona','cricket','sushant'])
```

```python
text= []
for i in all_tweets:
    text.append(i['text'])
hashtags = []
for i in text:
    for word in i.split():
        if "#" in word:
            hashtags.append(word)
```
