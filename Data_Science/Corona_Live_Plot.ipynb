```python
from matplotlib.animation import FuncAnimation
```

```python
dates = list(temp['ObservationDate'][:10])
v = list(temp['Confirmed'][:10])
fig = plt.figure(figsize=(9,5),facecolor="black")
ax = fig.add_axes([0,0,1,1],facecolor="black")
ax.plot(dates[:10],v[:10],'ro--')
#ax.tick_params(labelcolor="blue")
dates1 = list(temp['ObservationDate'][11:])
v1 = list(temp['Confirmed'][11:])
def change_plot(wait=None):
    global dates1,v1,dates,v
    dates.pop(0)  #to delete the value through which we have to plot new plot
    v.pop(0)  
    dates.append(dates1.pop(0))
    v.append(v1.pop(0))
    #sleep(wait)
    ax.clear()
    ax.plot(dates,v,'ro--')
    
f = FuncAnimation(fig,change_plot)
plt.show()
```
