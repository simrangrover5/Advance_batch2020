```python
df['Country/Region'] = df['Country/Region'].apply(lambda x : x.lower())
```

```python
df[df['ObservationDate'].apply(lambda x : True if int(x.split("/")[0]) > 3 else False)]
```

```python
fig = plt.figure(figsize=(11,5),facecolor="black")
ax = fig.add_axes([0,0,1,1],facecolor="black")
country = input("\n Enter any country : ")
mont = int(input("\n Enter the month : "))
temp = df[df['Country/Region'] == country]
v = temp[temp['ObservationDate'].apply(lambda x : True if x.split("/")[0] == ("0"+str(mont)) else False)]['Deaths'].values
i = temp[temp['ObservationDate'].apply(lambda x : True if x.split("/")[0] == ("0"+str(mont)) else False)]['ObservationDate'].values
ax.bar(i,v)
ax.set_xticklabels(i,rotation=90,color="white")
ax.set_yticklabels(ax.get_yticks(),color="white")
ax.clear()
ax.pie(v,labels=i,textprops={"color":"white"})
plt.show()
```

```python
fig = plt.figure(figsize=(11,5),facecolor="black")
ax = fig.add_axes([0,0,1,1],facecolor="black")
country = input("\n Enter any country : ")
mont = int(input("\n Enter the month : "))
temp = df[df['Country/Region'] == country]
v = temp[temp['ObservationDate'].apply(lambda x : True if x.split("/")[0] == ("0"+str(mont)) else False)]['Recovered'].values
i = temp[temp['ObservationDate'].apply(lambda x : True if x.split("/")[0] == ("0"+str(mont)) else False)]['ObservationDate'].values
ax.bar(i,v)
ax.set_xticklabels(i,rotation=90,color="white")
ax.set_yticklabels(ax.get_yticks(),color="white")
plt.savefig("recovered.png")
plt.show()
```
