```python
df1["C"].mean()
```

```python
def remove_out(df):
    value = 100
    for i in df:  #i --> columns
        for j in range(len(df[i])): #j --> values of each series
            v = df.loc[j,i]
            if v>100:
                df.loc[j,i] = np.nan
    return df
```
