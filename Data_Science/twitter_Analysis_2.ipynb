```python
def filter_hashtags(hashtags):
    newhash = []
    for i in hashtags:
        if i[0] == "#":
            if i[1:].isalpha():
                newhash.append(i)
    return newhash
```

```python
newhash = filter_hashtags(hashtags)
```

```python
fig = plt.figure(facecolor="#123456")
ax = fig.add_axes([0,0,1,1],facecolor="#123456")
sns.countplot(newhash,ax=ax)
ax.tick_params(axis="x",rotation=90)
plt.show()
```

```python
f = open("t.txt")
t = f.read().split("\n")
tweet = []
for i in t:
    try:
        tweet.append(json.loads(i))
    except:
        pass
```

```python
len(tweet)
```

```python
texts = []
for i in tweet:
    texts.append(i['text'])
```

```python
texts
```

```python
f = filter_text(texts)
```

```python
len(f)
```

```python
p = 0
neg = 0
n = 0
for i in f:
    a = TextBlob(i)
    if a.sentiment.polarity == 0:
        n += 1
    elif a.sentiment.polarity > 0:
        p += 1
    else:
        neg += 1
print(p,neg,n)
print("POSITIVE : ",(p/19)*100,"%")
print("NEGATIVE : ",(neg/19)*100,"%")
print("NEUTRAL :  ",(n/19)*100,"%")
```
