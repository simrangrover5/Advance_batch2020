```python
fig = plt.figure(facecolor="black")
ax1 = fig.add_axes([0,0,0.4,0.4],facecolor="black")
ax2 = fig.add_axes([0.5,0.5,0.4,0.4],facecolor="black")
sns.scatterplot("sepal_length","petal_length",data=iris,hue="species",ax=ax1)
sns.countplot("size",data=tips,ax=ax2,hue="sex",palette="rainbow",alpha=0.7)
ax1.set_title("SCATTERPLOT",fontdict={"fontsize":20,"color":"white"})
ax2.set_title("COUNTPLOT",fontdict={"fontsize":20,"color":"white"})
ax1.set_xlabel("SEPAL_LENGTH",fontdict={"fontsize":15,"color":"red"})
ax1.set_ylabel("PETAL_LENGTH",fontdict={"fontsize":15,"color":"red"})
ax2.set_xlabel("SIZE",fontdict={"fontsize":15,"color":"red"})
ax2.set_ylabel("COUNT",fontdict={"fontsize":15,"color":"red"})
ax1.grid()
ax2.grid()
plt.show()
```
