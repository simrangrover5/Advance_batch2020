```python
n = np.linspace(-5,5,20)
fig = plt.figure(dpi=90)
ax1 = fig.add_axes([0,0.5,0.3,0.5],facecolor="#123456")
ax2 = fig.add_axes([0.7,0.5,0.3,0.5],facecolor="#abcdef")
ax1.plot(n,n,'kh--',label="$X$")
ax2.plot(n,n**2,'k^:',label="$X^2$")
ax3 = fig.add_axes([0.35,0,0.3,0.5],facecolor="black")

ax3.plot(n,n**3,'o--',color="white",label="$X^3$")

for i in [ax1,ax2,ax3]:
    i.legend(loc="center")
    for sp in ['left','right','bottom','top']:
        i.spines[sp].set_edgecolor("red")
plt.show()
```

```python
fig = plt.figure(figsize=(11,5),facecolor="black")
ax1 = fig.add_axes([0,0.5,0.4,0.4],facecolor="black")
ax2 = fig.add_axes([0.5,0.5,0.4,0.4],facecolor="black")
ax3 = fig.add_axes([0.2,0,0.4,0.4],facecolor="black")
ax1.plot(x,sin,"r--",label="SIN(X)",lw=2)
ax2.plot(x,cos,"g:",label="COS(X)",lw=3)
ax3.plot(x,tan,'m-.',label="TAN(X)",lw=3)
for ax in [ax1,ax2,ax3]:
    ax.tick_params(labelcolor="white")
    ax.grid()
    ax.legend()
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(11,5))
for r in range(len(ax)):
    for c in range(len(ax[0])):
        if r == c:
            ax[r][c].plot(x,x**2,'r--')
        else:
            ax[r][c].plot(x,x**4,'b--')
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(11,5))
ax[0][0].plot(x,sin,'r--')
ax[0][1].plot(x,cos,'g--')
ax[1][1].plot(x,tan,'m--')
ax[1][0].plot(x,np.cosh(x),'k--')
for i in range(len(ax)):
    for c in range(len(ax[0])):
        ax[i][c].spines["left"].set_edgecolor("yellow")
        ax[i][c].spines["right"].set_edgecolor("#123456")
        ax[i][c].spines["top"].set_edgecolor("#abcdef")
        ax[i][c].spines["bottom"].set_edgecolor("#1234ab")
        ax[i][c].spines["left"].set_linewidth(5)
        ax[i][c].spines["bottom"].set_linewidth(5)
        if i == c:
            ax[i][c].tick_params(labelcolor="purple",labelsize=15)
        else:
            ax[i][c].tick_params(labelcolor="red",labelsize=15)
plt.show()
```
