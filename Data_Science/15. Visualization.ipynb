```python
fig = plt.figure(dpi=90,facecolor="black")
ax1 = fig.add_axes([0,0,0.5,0.5],facecolor="black")
ax2 = fig.add_axes([0.5,0.5,0.5,0.5],facecolor="black")
n = np.linspace(-5,5,20)
ax1.plot(n,n**2,'*--')
ax2.plot(n,n**3,"go--")
ax1.set_xticklabels(ax1.get_xticks(),color="red")
ax2.set_xticklabels(ax2.get_xticks(),color="red")
ax1.set_yticklabels(ax1.get_yticks(),color="red")
ax2.set_yticklabels(ax2.get_yticks(),color="red")
for i in [ax1,ax2]:
    for sp in ["left","right","top","bottom"]:
        i.spines[sp].set_edgecolor("white")
plt.show()
```

```python
fig = plt.figure(dpi=90)
ax1 = fig.add_axes([0,0.5,0.5,0.5],facecolor="#123456")
ax2 = fig.add_axes([0.5,0,0.5,0.5],facecolor="#abcdef")
ax1.plot(n,n**2,'kh--')
ax2.plot(n,n**3,'k^:')
plt.show()
```

```python
fig = plt.figure(dpi=90,facecolor="black")
ax1 = fig.add_axes([0,0.5,0.5,0.5],facecolor="#123456")
ax2 = fig.add_axes([0.5,0,0.5,0.5],facecolor="#abcdef")
ax1.plot(n,n,'kh--',label="$X$")
ax2.plot(n,n**2,'k^:',label="$X^2$")
ax3 = fig.add_axes([0,0,0.5,0.5],facecolor="black")
ax4 = fig.add_axes([0.5,0.5,0.5,0.5],facecolor="black")
ax3.plot(n,n**3,'o--',color="white",label="$X^3$")
ax4.plot(n,n**4,'p-.',color="white",label="$X^4$")
for i in [ax1,ax2,ax3,ax4]:
    i.set_xticks([])
    i.set_yticks([])
    i.legend(loc="center")
    for sp in ['left','right','bottom','top']:
        i.spines[sp].set_edgecolor("red")
plt.show()
```
