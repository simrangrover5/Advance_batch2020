```python
from time import sleep

x = pd.date_range("20200116",periods=10)
y = list(np.random.randint(10,60,10))
fig,ax = plt.subplots()
ax.plot(x,y,'ro--')

def animate(wait=None):
    global y
    ax.clear()
    sleep(wait)
    y.pop(0)
    y.append(np.random.randint(10,60))
    ax.plot(x,y,'ro--')

f = FuncAnimation(fig,animate,2)
plt.show()
```

```python
x = list(range(50))
y = [psutil.cpu_percent() for i in range(50)]
fig = plt.figure(figsize=(10,5),facecolor="black")
ax = fig.add_axes([0,0,1,1],facecolor="black")
ax.plot(x,y,'ro--')
def animation(wait=None):
    global y
    ax.clear()
    y.pop(0)
    sleep(wait)
    y.append(psutil.cpu_percent())
    ax.plot(x,y,'wo--')
    
f = FuncAnimation(fig,animation,2)
plt.show()
```

```python
x = list(range(50))
y = [psutil.cpu_percent() for i in range(50)]
fig = plt.figure(figsize=(9,5),facecolor="black")
ax = fig.add_axes([0,0,1,1],facecolor="black")
ax.scatter(x,y)
ax.tick_params(axis="y",labelcolor="red")
def animation(wait=None):
    global y
    ax.clear()
    y.pop(0)
    sleep(wait)
    y.append(psutil.cpu_percent())
    ax.scatter(x,y)
    ax.tick_params(axis="y",labelcolor="red")
    
f = FuncAnimation(fig,animation,2)
plt.show()
```
