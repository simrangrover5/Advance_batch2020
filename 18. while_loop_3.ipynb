```python
#palindrome
n = int(input("\n Enter any number : "))
s = 0
n1 = n  #bcz we are modifying the value of n at each iteration so to get the actual value at last to match we will store it into some temporary variable
while n>0:
    r = n%10 #it will give us the reminder that means the last number
    #at each iteration by using % operator we will have the last value
    s = s*10 + r
    n = n//10  #at each iteration it will give me the value of n by removing the last digit

if s == n1:
    print(f"\n {n1} is palindrome ")
else:
    print(f"\n {n1} is not palindrome")

#i. n = 121 , r = 1, s = 0*10+1 = 0+1 = 1, n=12
#ii. n = 12, r = 2 , s = 1*10 + 2 = 10+12 = 12, n = 1
#iii n = 1 ,r = 1, s = 12*10 + 1 = 120 + 1 = 121, n = 0
```

```python
colors = ['red','blue','green','orange','purple','yellow']
i = 0
while i<len(colors):
    c = input("\n Enter a color : ")
    if "r" in c:
        print("\n I don't want to continue as i don't like r")
        break
    print("\n I like ",c)
    i += 1
else:
    print("\n Oh welcome you entered the else part.....")
```
