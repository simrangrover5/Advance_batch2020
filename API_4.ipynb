```python
import requests

url = "https://covid-193.p.rapidapi.com/statistics"

headers = {
    'x-rapidapi-host': "covid-193.p.rapidapi.com",
    'x-rapidapi-key': "a88ff77402mshb59de67584167e6p113a98jsncad39a9d4c55"
    }

response = requests.request("GET", url, headers=headers)

data = response.json()
```

```python
country = input("Enter any country : ").lower()
for i in data['response']:
    if i['country'].lower() == country:
        print("#"*80)
        print("Total Population : ",i['population'])
        print("New cases : ",i['cases']['new'])
        print("Active : ",i['cases']['active'])
        print("Recovered : ",i['cases']['recovered'])
        print("Total Cases : ",i['cases']['total'])
        print("Total deaths : ",i['deaths']['total'])
        print("New deaths : ",i['deaths']['new'])
        print("Test Cases : ",i['tests']['total'])
        print("Day : ",i['day'])
        print("Time : ",i['time'])
        print("#"*80)
```
