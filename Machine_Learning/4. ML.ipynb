#### Error
    The difference between the actual value and predicted value is error
    
$$ error = y_i - \hat y_i $$

    1. Bestfit
        The line which covers almost every point and different behaviour of point
        The error in this case is less and accuracy is more
    2. Underfit
        The most of the points are not covered due to which the error is more and accuracy is less
    3. Overfit
        All the training points are convered which becomes sentisitive for the points outside the
        training points due to which error is more and accuracy is less

$$ y = \beta_0 + \beta_1 x $$

$ \beta_0 $ is intercept
$ \beta_1 $ is coeff or slope

### Mean Absolute Error

$$ mae = \frac {\sum\limits_{i=0}^N y_i - \hat y _i} {N} $$

## Mean Square Error


$$ MSE = \frac {\sum\limits_{i=0}^N (y_i - \hat y_i)^2} {N} $$

$$ y = \beta_0 + \beta_1 x $$

Replace the value of $ y_i $ and $ \hat y_i $ in above mse equation then the following equation results

$$ \text{MSE} = \frac {\sum\limits_{i=0}^N {((\beta_0 + \beta_1 x_i) - (\beta_0 + \beta_1 x))}^2} {N} $$

Differentiate this equation w.r.t $ \beta_0 $ and $ \beta_1 $ and after that the following equation results

$$ \beta_1 = \frac {\sum\limits_{i=0}^N (x_i -\bar x_i)(y_i - \bar y_i)} {\sum\limits_{i=0}^N(x_i - \bar x_i)^2} $$

$$ \beta_0 = \bar y - \beta_1 \bar x $$

## Root Mean Square Error


$$ \text{RMSE} = \sqrt {\frac {\sum\limits_{i=0}^N (y_i - \hat y_i)^2} {N}} $$

### Accuracy

$$ \text{r2_score} = 1 - \frac {mse} {\sum\limits_{i=0}^N (y_i - \hat y_i)^2}  $$

    The value of r_2 score is between 0 to 1
    
    0 --> Error is way more
    So this is the worst case 
    
    1 --> This is best case and it means your model is working perfectly
