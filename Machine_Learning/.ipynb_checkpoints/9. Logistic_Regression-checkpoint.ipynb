    but the data which it handles is categorial data
     
    This algorithm is for the classification
    
    
    
    Categorial Data 
    
    Examples --> Gender (Male, Female) , girl and boy, andoid and ios, yes and no, cbse and state board, 
    fruits and vegetables, 0 and 1 , summer winter and autumn , types of transport, colors, 4 wheeler and 2 wheeler, 
    good better avg and best , language, fruad and non fruad, happy sad and neutral etc
