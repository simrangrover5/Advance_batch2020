### Linear Regression

        Regression
            Predict dependent variables/values on basis of independent variables
            
         Linear Regression
             Increasing one variable makes increase in another also
             Increasing one variables makes another deacrease
             

    The line which you will use for prediction is found using this equation
    
    y = mx + c
    
    y --> target/dependent variable
    x --> feature/independent variable 
    m --> slope
    c --> intercept
    
    error --> actual value - predicted value

    To find the values of m and c
    
    1. Check arbitary values of m and c and find best fit lines
    2. Use formula to find the values of m and c

$$ x^2 + y^2 $$

$ x^2 + y^2 $

$$ y_i = x_i + c $$

$$ \text{Equation for line} = \beta_0 + \beta_1*x $$

$$ y = \beta_0 + \beta_1x $$

where $\beta_0$ is intercept and $\beta_1$ is slope 

$$ \alpha, \beta, \gamma $$

$$ a = \frac {x^2 + y^2} {x^2 - y^2} $$

$$ \tan = \frac {\sin} {\cos} $$

$$ \log a $$

$$ mean = \frac {\sum\limits_{i=0}^N x_i} {N} $$ 

$$ a = \sqrt{x^2 + y^2} $$
