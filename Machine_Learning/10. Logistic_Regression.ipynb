    2 categories --> Binary Classification
    Yes and no, male and female, good and bad
    
    Multiple Categories --> Multivariant Classification
    iris --> setosa, verginica, versicolor 

    Logistics
    
    Sigmoid or Logit function --> S shape curve (which is used for prediction) 
    
$$ \text {Probability}     =        \frac {\text {favourable outcomes}} {\text{total outcome}} $$    

$$ P(X)    = \frac {p(x)} {1 - p(x)} $$


    
    x = 1
    p(1) = Probability of credit card being fraud
    1 - p(1) = Probability of credit card nor fraud
    
    
$$ P(X) = \beta_0 + \beta_1 x $$

$$ \beta_0 + \beta_1 x = \frac {p(x)} {1 - p(x)} $$

$$ logit function --> log $$

$$ \beta_0 + \beta_1 x = \log {\frac {p(x)} {1 - p(x)}} $$


$$ \frac {p(x)} {1 - p(x)} = e ^{ (\beta_0 + \beta_1 x)} $$


$$ p(x) = e ^{ (\beta_0 + \beta_1 x)} * (1 - p(x)) $$

$$ p(x) = e ^{ (\beta_0 + \beta_1 x)} - e ^{ (\beta_0 + \beta_1 x)}p(x) $$

$$ p(x) + e ^{ (\beta_0 + \beta_1 x)}p(x) = e ^{ (\beta_0 + \beta_1 x)} $$

$$ p(x) (1 + e ^{ (\beta_0 + \beta_1 x)} ) = e ^{ (\beta_0 + \beta_1 x)} $$

$$ f(x) = \frac {1} {1 + e^{-z}} $$

where f(x) is logit function and z is $ \beta_0 + \beta_1 x $

```python
features = data[['balance','income']]
target = pd.Series(data['student']).map({"No":0,"Yes":1})
```

```python
pred1 = pred.copy()
pred1[pred1>thres] = 1
pred1[pred1<thres] = 0
plt.plot(features['income'],target,'o')
plt.plot(features['income'],pred1,'ko')
plt.show()
```

```python
accuracy_score(target,pred1)
```

```python
for thres in np.arange(0.3,0.8,0.1):
    pred1 = pred.copy()
    pred1[pred1>thres] = 1
    pred1[pred1<thres] = 0
    plt.plot(features['income'],target,'o')
    plt.plot(features['income'],pred1,'ko')
    plt.title(f"{thres} --> {accuracy_score(target,pred1)}")
    plt.show()
```
